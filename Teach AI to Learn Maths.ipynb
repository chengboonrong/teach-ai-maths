```python
import random
numbers = [i for i in range(101)]
```

```python
def add(l):
    a = l[random.randint(0,len(l)-1)]
    b = l[random.randint(0,len(l)-1)]
    c = a + b
    return a, b, c
```

```python
x = []
target = []

for i in range(1000):
    n1, n2, result = add(numbers)
    x.append([n1,n2])
    target.append(result)    
```
